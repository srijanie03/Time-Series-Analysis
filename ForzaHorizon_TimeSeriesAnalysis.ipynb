```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
forza4.set_index('DateTime', inplace=True)
analysis4 = forza4[['Players']].copy()
forza5.set_index('DateTime', inplace=True)
analysis5 = forza5[['Players']].copy()
```

```python
df_forza4
```

```python
forza4
```

```python
decompose_result_mult4 = seasonal_decompose(analysis4, model="multiplicative")
decompose_result_mult5 = seasonal_decompose(analysis5, model="multiplicative")

trend4 = decompose_result_mult4.trend
seasonal4 = decompose_result_mult4.seasonal
residual4 = decompose_result_mult4.resid
decompose_result_mult4.plot()

trend5 = decompose_result_mult5.trend
seasonal5 = decompose_result_mult5.seasonal
residual5 = decompose_result_mult5.resid
```

```python
decompose_result_mult5.plot()
```

```python
plt.figure(figsize=(15,4))
plt.plot(x4[200:300],seasonal4[200:300],label='Horizon4')
plt.plot(x5[200:300],seasonal5[200:300],label='Horizon5')
plt.legend()
plt.grid()
plt.show()
```

```python
fig,ax =  plt.subplots(2,figsize=(15, 8))
ax[0].plot(x4[463:490],seasonal4[463:490],'darkblue',label='Horizon4') #450:550
ax[0].legend()
ax[0].grid(color='purple',ls=':',alpha=0.5)
ax[1].plot(x5[220:245],seasonal5[220:245],'magenta',label='Horizon5') #200:300
ax[1].legend()
ax[1].grid(color='purple',ls=':',alpha=0.5)
ax[0].set_title('Seasonal Variation')
for axi in ax.flat:
    axi.set(xlabel='Time', ylabel='Seasonal(Daily CCU)')
plt.show()
```
